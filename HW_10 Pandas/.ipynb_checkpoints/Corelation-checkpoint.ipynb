```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv',header=0)
```

```python
df.head()
```

```python
cols = ['Name','Ticket','Cabin', 'PassengerId']
df = df.drop(cols,axis=1)
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df['Sex'] = df['Sex'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')

df['Sex_codes'] = df['Sex'].cat.codes
df['Pclass_codes'] = df['Pclass'].cat.codes

embarked_coding = {"Embarked": {'S': 0, 'C': 1, 'Q': 2}}
df.replace(embarked_coding, inplace=True)
#df['Embarked'] = df['Embarked'].astype('Embarked')
```

```python
cols = ['Sex','Pclass']
df = df.drop(cols,axis=1)
```

```python
df.head()
```

```python
sns.set(font_scale=1)
fig = plt.figure(figsize=(15, 10))
hm = sns.heatmap(df.corr(), cbar=True, annot=True, fmt='.2f', annot_kws={'size': 15}, yticklabels=True)
hm.set_yticklabels(hm.get_yticklabels(), rotation=0)
plt.title('') 
plt.show()
```

```python
#pd.plotting.scatter_matrix(df, figsize=(30, 20), diagonal='kde')
axes = pd.plotting.scatter_matrix(df, alpha=0.8, figsize=(25, 20), diagonal='kde')
for ax in axes.flatten():
    ax.xaxis.label.set_rotation(0)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')

plt.tight_layout()
plt.gcf().subplots_adjust(wspace=0, hspace=0)
plt.show()
```
